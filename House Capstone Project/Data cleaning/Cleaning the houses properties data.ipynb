```python
##If there are basically no data within a specific columns, we should just remove it

def drop_useless_columns(dataframe):
    columns_unique = dataframe.columns.tolist()
    columns_to_drop = list()
    rows_check = (dataframe.shape[0]) * 0.90
    for i in range(len(columns_unique)):
        column = columns_unique[i]
        if (dataframe.loc[dataframe[column].isnull()].shape[0] >= rows_check) or (df[column].nunique() <= 1):
            columns_to_drop.append(column)
    columns_chosen = list(set(columns_unique) - set(columns_to_drop))
    #print("Drop these columns: ", columns_to_drop)
    return dataframe[columns_chosen]
```
