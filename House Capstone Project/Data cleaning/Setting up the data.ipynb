## Testing out the api function with one city and within the price range
## 200,001 to 300,000 in (US dollars)

```python
#Test api with fortworth
#This first block runs the first page
#And this whole cell is only for 1 city with this price range

all_responses = list()
url = "https://zillow-com1.p.rapidapi.com/propertyExtendedSearch"
querystring = {"location":search_str[0], "home_type":"Houses", "minPrice":200001, "maxPrice":300000}
headers = {
	"X-RapidAPI-Host": "zillow-com1.p.rapidapi.com",
	"X-RapidAPI-Key": api_key
}
response = requests.request("GET", url, headers=headers, params=querystring)
fort_worth_2_3 = response.json()
all_responses.append(fort_worth_2_3)
total_pages = fort_worth_2_3['totalPages']

##Wait 10 seconds due to api call
time.sleep(10)

##If total pages > 1
if total_pages > 1:
    count = 1
    for i in range(1, total_pages):
        querystring = {"location":search_str[0], "page": count + i, "home_type":"Houses", 
                       "minPrice":200001, "maxPrice":300000}
        headers = {
                    "X-RapidAPI-Host": "zillow-com1.p.rapidapi.com",
                    "X-RapidAPI-Key": api_key}
        response = requests.request("GET", url, headers=headers, params=querystring)
        fort_worth_2_3 = response.json()
        all_responses.append(fort_worth_2_3)
        time.sleep(10)
```

```python
all_dfs = list()
for i in range(len(all_responses)):
    df = pd.json_normalize(data = all_responses[i]['props'])
    all_dfs.append(df)
```

```python
fw_df = pd.concat(all_dfs, axis = 0)
fw_df.reset_index(drop= True, inplace = True)
fw_df
```
